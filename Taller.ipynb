```python
import math

# 1. Cálculo de la Caída Libre
def caida_libre():
    h = float(input("Ingrese la altura en metros: "))
    g = 9.81
    t = math.sqrt((2 * h) / g)
    print(f"El tiempo de caída es: {t:.2f} segundos\n")


# 2. Conversión de Unidades de Velocidad
def conversion_velocidad():
    v = float(input("Ingrese el valor de la velocidad: "))
    tipo = input("Escriba '1' para km/h a m/s o '2' para m/s a km/h: ")

    if tipo == "1":
        resultado = v / 3.6
        print(f"{v} km/h = {resultado:.2f} m/s\n")
    elif tipo == "2":
        resultado = v * 3.6
        print(f"{v} m/s = {resultado:.2f} km/h\n")
    else:
        print("Opción inválida\n")


# 3. Cálculo del Desplazamiento
def desplazamiento():
    u = float(input("Ingrese la velocidad inicial (m/s): "))
    a = float(input("Ingrese la aceleración (m/s^2): "))
    t = float(input("Ingrese el tiempo (s): "))

    s = u * t + 0.5 * a * t**2
    print(f"El desplazamiento es: {s:.2f} metros\n")


# 4. Suma de Vectores
def suma_vectores():
    x1 = float(input("Ingrese x1: "))
    y1 = float(input("Ingrese y1: "))
    x2 = float(input("Ingrese x2: "))
    y2 = float(input("Ingrese y2: "))

    suma = [x1 + x2, y1 + y2]
    print(f"La suma vectorial es: {suma}\n")


# 5. Producto Escalar y Ángulo entre Vectores
def producto_escalar():
    x1 = float(input("Ingrese x1: "))
    y1 = float(input("Ingrese y1: "))
    x2 = float(input("Ingrese x2: "))
    y2 = float(input("Ingrese y2: "))

    producto = x1 * x2 + y1 * y2
    magnitud1 = math.sqrt(x1**2 + y1**2)
    magnitud2 = math.sqrt(x2**2 + y2**2)

    if magnitud1 == 0 or magnitud2 == 0:
        print("No se puede calcular el ángulo con un vector nulo.\n")
        return

    cos_theta = producto / (magnitud1 * magnitud2)
    # Ajuste por posibles errores numéricos
    cos_theta = max(min(cos_theta, 1), -1)

    angulo = math.degrees(math.acos(cos_theta))

    print(f"Producto escalar: {producto}")
    print(f"Ángulo entre vectores: {angulo:.2f} grados\n")


# 6. Lanzamiento de Proyectil
def lanzamiento_proyectil():
    v0 = float(input("Ingrese la velocidad inicial (m/s): "))
    theta = float(input("Ingrese el ángulo (grados): "))
    g = 9.81

    theta_rad = math.radians(theta)

    R = (v0**2 * math.sin(2 * theta_rad)) / g
    H = (v0**2 * math.sin(theta_rad)**2) / (2 * g)

    print(f"Alcance máximo: {R:.2f} metros")
    print(f"Altura máxima: {H:.2f} metros\n")


# Menú Principal
def menu():
    while True:
        print("===== MENÚ DE OPCIONES =====")
        print("1. Cálculo de la Caída Libre")
        print("2. Conversión de Unidades de Velocidad")
        print("3. Cálculo del Desplazamiento")
        print("4. Suma de Vectores")
        print("5. Producto Escalar y Ángulo entre Vectores")
        print("6. Lanzamiento de Proyectil")
        print("7. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            caida_libre()
        elif opcion == "2":
            conversion_velocidad()
        elif opcion == "3":
            desplazamiento()
        elif opcion == "4":
            suma_vectores()
        elif opcion == "5":
            producto_escalar()
        elif opcion == "6":
            lanzamiento_proyectil()
        elif opcion == "7":
            print("Saliendo del programa...")
            break
        else:
            print("Opción inválida. Intente nuevamente.\n")


# Ejecutar programa
menu()
```
